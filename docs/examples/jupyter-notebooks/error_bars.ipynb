```kotlin
%use lets-plot
```

### Plotting means and error ranges.

There are several ways to show error ranges on a plot. Among them are
- *geom_errorbar*
- *geom_crossbar*
- *geom_linerange*
- *geom_pointrange*

```kotlin
// This example was found at: www.cookbook-r.com/Graphs/Plotting_means_and_error_bars_(ggplot2)
val data = mapOf(
    "supp" to listOf("OJ", "OJ", "OJ", "VC", "VC", "VC"),
    "dose" to listOf(0.5, 1.0, 2.0, 0.5, 1.0, 2.0),
    "length" to listOf(13.23, 22.70, 26.06, 7.98, 16.77, 26.14),
    "len_min" to listOf(11.83, 21.2, 24.50, 4.24, 15.26, 23.35),
    "len_max" to listOf(15.63, 24.9, 27.11, 10.72, 19.28, 28.93)
)
```

```kotlin
val p = lets_plot(data) {x="dose"; color="supp"}
```

### Error-bars with lines and points.

```kotlin
p + geom_errorbar(width=.1) {ymin="len_min"; ymax="len_max"} +
    geom_line {y="length"} +
    geom_point {y="length"}
```

```kotlin
// The errorbars overlapped, so use position_dodge to move them horizontally
val pd = position_dodge(0.1) // move them .05 to the left and right
p + geom_errorbar(width=.1, position=pd) {ymin="len_min"; ymax="len_max"} +
    geom_line(position=pd) {y="length"} +
    geom_point(position=pd) {y="length"}
```

```kotlin
// Black errorbars - notice the mapping of 'group=supp'
// Without it, the errorbars won't be dodged!
p + geom_errorbar(color="black", width=.1, position=pd) {ymin="len_min"; ymax="len_max"; group="supp"} +
    geom_line(position=pd) {y="length"} +
    geom_point(position=pd, size=5.0) {y="length"}
```

```kotlin
// Finished graph
// - fixed size
// - point shape # 21 is filled circle 
// - position legend in bottom right
val p1 = p +
    xlab("Dose (mg)") +
    ylab("Tooth length (mm)") +
    scale_color_manual(listOf("orange", "dark_green"), naValue="gray") +
    ggsize(700, 400)
p1 + geom_errorbar(color="black", width=.1, position=pd) {ymin="len_min"; ymax="len_max"; group="supp"} +
    geom_line(position=pd) {y="length"} +
    geom_point(position=pd, size=5.0, shape=21, fill="white") {y="length"} +
    theme().legendJustification(1,0).legendPosition(1,0) +
    ggtitle("The Effect of Vitamin C on Tooth Growth in Guinea Pigs")
```

### Error-bars on bar plot.

```kotlin
// Plot error ranges on Bar plot
p1 + 
    geom_bar(stat=Stat.identity, position=Pos.dodge, color="black") {y="length"; fill="supp"} +
    geom_errorbar(color="black", width=.1, position=position_dodge(0.9)) {
        ymin="len_min"
        ymax="len_max" 
        group="supp"} +
    theme().legendJustification(0,1).legendPosition(0,1)
```

### Crossbars.

```kotlin
// Thickness of the horizontal mid-line can be adjusted using `fatten` parameter.
p1 + geom_crossbar(fatten=5.0) {
    ymin="len_min"
    ymax="len_max"
    middle="length"
    color="supp"
}
```

### Line-range.

```kotlin
p1 + geom_linerange(position=pd) {ymin="len_min"; ymax="len_max"; color="supp"} +
     geom_line(position=pd) {y="length"}
```

### Point-range.

```kotlin
// Point-range is the same as line-range but with an added mid-point.
p1 + geom_pointrange(position=pd) {y="length"; ymin="len_min"; ymax="len_max"; color="supp"} +
     geom_line(position=pd) {y="length"}
```

```kotlin
// Size of the mid-point can be adjuasted using `fatten` parameter - multiplication factor relative to the line size.
p1 + geom_line(position=pd) {y="length"} +
     geom_pointrange(position=pd, color="rgb(230, 230, 230)", size=5.0, shape=23, fatten=1.0) {
         y="length"
         ymin="len_min"
         ymax="len_max"
         fill="supp"
     } +
     scale_fill_manual(listOf("orange", "dark_green"), naValue="gray")
```
